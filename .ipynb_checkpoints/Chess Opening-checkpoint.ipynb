```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

Dataset from Kaggle: [Chess Game Dataset(Lichess)](https://www.kaggle.com/datasnaek/chess)

```python
df = pd.read_csv('games.csv')
df.head()
```

``opening_archetype``: Overall "type" of a opening that is used (Information on opening game is included in ``opening_name``.)

```python
df = (df.assign(opening_archetype=df['opening_name'].map(lambda n: n.split(":")[0].split("|")[0].split("#")[0].strip()), 
                opening_moves=df.apply(lambda srs: srs['moves'].split(" ")[:srs['opening_ply']], axis=1)))
```

Find players with enough data, i.e., players who've played more than two games.

```python
games_played = pd.concat([df['white_id'], df['black_id']]).value_counts()
games_played.reset_index(drop=True).plot.line(figsize=(16, 8), fontsize=18)
ngt = len(games_played[games_played>1])
print(ngt," players have played at least two games.")
print("Games played by", ngt, "players:", games_played[games_played>1].sum())
plt.axvline(ngt, color='green')
```

```python
opening_used = (pd.concat([df.groupby('white_id')['opening_archetype'].value_counts(), 
                           df.groupby('black_id')['opening_archetype'].value_counts()]))
opening_used.head()
```

```python
opening_used = opening_used.rename(index='openings_used').reset_index()
```

```python
opening_used = opening_used.rename(columns={'white_id': 'player_id', 'openings_used': 'times_used'})
opening_used = opening_used.groupby(['player_id', 'opening_archetype']).sum()
opening_used.head()
```

```python
opening_used.shape
```

Most commonly used chess openings are **Sicilian Defense** and **King's Gambit**. The strong preference of chess players towards the same opening move would make any recommendation less interesting.

```python
sicilian = (opening_used
            .reset_index()
            .groupby('player_id')
            .filter(lambda g: g.opening_archetype.isin(['Sicilian Defense']).any())
            .query('opening_archetype != "Sicilian Defense"')
            .groupby('opening_archetype')
            .times_used
            .sum()
            .sort_values(ascending=False)
            .to_frame()
            .pipe(lambda g: g.assign(times_used = g.times_used/g.times_used.sum()))
            .squeeze())
sicilian.head(10)
```

```python
kingsg = (opening_used
          .reset_index()
          .groupby('player_id')
          .filter(lambda g: g.opening_archetype.isin(["King's Gambit"]).any())
          .query('opening_archetype != "King\'s Gambit"')
          .groupby('opening_archetype')
          .times_used
          .sum()
          .sort_values(ascending=False)
          .to_frame()
          .pipe(lambda g: g.assign(times_used = g.times_used/g.times_used.sum()))
          .squeeze())
kingsg.head(10)
```

* A King's Pawn Game is ~50% more likely for a King's Gambit player than it is for a Sicilian Defense player.
* English Opening is much less popular with the Sicilian Defense player.
* Other differences are also apparent.

Variation in chess openings seems significant which provides decent evidence that recommendations for chess openings would be useful.

### Recommendation engine using collaborative filtering

The dataset doesn't have rating for openings. Ratings can be assigned to openings used by a player using simple exponential threshold approach. If a player has played ``N`` games and used the given opening for >``N/4`` times, then that open is ⭐⭐⭐⭐⭐ for that player. Similarly, >``N/8`` is ⭐⭐⭐⭐ and >``N/16`` is ⭐⭐⭐.

```python
def threshold_map(n_opening, n_all):
    if pd.isnull(n_opening):
        return np.nan
    elif n_opening/n_all >= 1/4:
        return 5
    elif n_opening/n_all >= 1/8:
        return 4
    elif n_opening/n_all > 1/16:
        return 3
    else:
        return 2
```

```python
recommendations = opening_used.unstack(-1).loc[:, 'times_used'].apply(
    lambda srs: srs.map(lambda v: threshold_map(v, srs.sum())),
    axis='columns')
recommendations.head()
```

```python
pd.Series(recommendations.values.flatten()).value_counts().sort_index().plot.bar()
```

Distribution is dominated by 5-star ratings.

```python
from sklearn.metrics import pairwise_distances
item_similarity = pairwise_distances(recommendations.T.fillna(0), metric='cosine')
item_similarity.shape
```

There are total 143 strategies.

```python
correction = np.array([np.abs(item_similarity).sum(axis=1)])
item_predictions = recommendations.fillna(0).dot(item_similarity).apply(
    lambda srs: srs/np.array([item_similarity.sum(axis=1)]).flatten(), axis='columns')
item_predictions.head()
```

```python
recommended_opening_nos = item_predictions.apply(lambda srs: np.argmax(srs.values), axis='columns')
recommended_opening_nos.head()
```

```python
opening_names = pd.Series(recommendations.columns)
recommended_openings = recommended_opening_nos.map(opening_names)
recommended_openings.head()
```

```python
recommended_openings.value_counts().iloc[1:].plot.bar(figsize=(24, 12), fontsize=12)
```

``System`` has been excluded because it seems more or less like a default option. It doesn't imply to any set methodology and doesn't sound like an interesting recommendation.
